```r
install.packages("quantmod")
library(quantmod)
```

```r
getSymbols("ZOMATO.NS", src = "yahoo")
```

```r
options(repr.plot.width = 14, repr.plot.height = 8)
chartSeries(ZOMATO.NS, theme = chartTheme("black"), type = "line")
```

```r
options(repr.plot.width = 14, repr.plot.height = 8)
chartSeries(ZOMATO.NS, theme = chartTheme("black"), TA ="addSMA(n=20,col='blue') ; addSMA(n=50,col='red')")
```

```r
print("Staring price of the stock")
head(ZOMATO.NS)
print("----------------------------------------------------")
print("Latest price of the stock")
tail(ZOMATO.NS)
```

```r
daily_returns <- dailyReturn(Cl(ZOMATO.NS))
plot(daily_returns, main = "Daily Returns of ZOMATO.NS")
```
